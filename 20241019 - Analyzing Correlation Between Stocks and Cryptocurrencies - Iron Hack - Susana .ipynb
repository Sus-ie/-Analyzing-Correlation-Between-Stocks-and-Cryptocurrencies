```python
import pandas as pd
from alpaca.data.historical import CryptoHistoricalDataClient
from alpaca.data.requests import CryptoBarsRequest
from alpaca.data.timeframe import TimeFrame

# ETH/USD - Ethereum / US Dollar pull every two weeks
client = CryptoHistoricalDataClient()

start_date = "2023-01-01"
end_date = "2023-12-31"

request_params = CryptoBarsRequest(
  symbol_or_symbols=["ETH/USD"],
  timeframe=TimeFrame.Week,
  start=start_date,
  end=end_date
)

crypto_bars = client.get_crypto_bars(request_params)

crypto_df = crypto_bars.df

crypto_df.index = pd.to_datetime(crypto_df.index.get_level_values(1))

bi_weekly_crypto_df_eth = crypto_df.resample('2W').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum',
    'trade_count': 'sum'
})

bi_weekly_crypto_df_eth['close_pct_change'] = bi_weekly_crypto_df_eth['close'].pct_change() * 100
```

```python
stock_symbols = ['AMZN', 'AAPL', 'MSFT', 'NVDA', 'META']

# Create a dictionary to df (stock)
bars_dfs = {}

# Fetch stock data and store in dictionray
for symbol in stock_symbols:
    request_params = StockBarsRequest(
        symbol_or_symbols=[symbol],
        timeframe=TimeFrame.Week,
        start=start_time,
        end=end_time
    )
    bars_df = data_client.get_stock_bars(request_params).df.tz_convert('America/New_York', level=1)
    bars_df.index = pd.to_datetime(bars_df.index.get_level_values(1))
    bars_dfs[symbol] = bars_df

# Assign stock df into a variable
bars_df_amzn = bars_dfs['AMZN']
bars_df_aapl = bars_dfs['AAPL']
bars_df_msft = bars_dfs['MSFT']
bars_df_nvda = bars_dfs['NVDA']
bars_df_meta = bars_dfs['META']

# Convert Stock to UTC to align with crypto data (Bitcoin and Ethereum). Issue here is that the stocks have set "opening times" for trading
# This being said crpyto has no "open and close times"
bars_df_amzn.index = bars_df_amzn.index.tz_convert('UTC')
bars_df_aapl.index = bars_df_aapl.index.tz_convert('UTC')
bars_df_msft.index = bars_df_msft.index.tz_convert('UTC')
bars_df_nvda.index = bars_df_nvda.index.tz_convert('UTC')
bars_df_meta.index = bars_df_meta.index.tz_convert('UTC')

# RECALCULATE the percentage change for each stock
bi_weekly_df_amzn = bars_df_amzn.resample('2W').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum',
    'trade_count': 'sum'
})
bi_weekly_df_aapl = bars_df_aapl.resample('2W').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum',
    'trade_count': 'sum'
})
bi_weekly_df_msft = bars_df_msft.resample('2W').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum',
    'trade_count': 'sum'
})
bi_weekly_df_nvda = bars_df_nvda.resample('2W').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum',
    'trade_count': 'sum'
})
bi_weekly_df_meta = bars_df_meta.resample('2W').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum',
    'trade_count': 'sum'
})

#please note that there is repetition in code because the plan is to simplify and reduced code lines in general thus a lot of repeition!
# Recalculate percentage change for all stocks (the purpose here was that we want to only have one long script so will try and clean up the previous script and keep only this part 
bi_weekly_df_amzn['close_pct_change'] = bi_weekly_df_amzn['close'].pct_change() * 100
bi_weekly_df_aapl['close_pct_change'] = bi_weekly_df_aapl['close'].pct_change() * 100
bi_weekly_df_msft['close_pct_change'] = bi_weekly_df_msft['close'].pct_change() * 100
bi_weekly_df_nvda['close_pct_change'] = bi_weekly_df_nvda['close'].pct_change() * 100
bi_weekly_df_meta['close_pct_change'] = bi_weekly_df_meta['close'].pct_change() * 100


# Again, example of repetition just to ensure that Bitcoin data is in UTC
bi_weekly_crypto_df_btc.index = bi_weekly_crypto_df_btc.index.tz_convert('UTC')  # Ensure Bitcoin data is in UTC

# Resample and calculate percentage change the "CLOSE" date Bitcoin
bi_weekly_crypto_df_btc['close_pct_change'] = bi_weekly_crypto_df_btc['close'].pct_change() * 100

# Function to calculate Pearson correlation for a stock vs Bitcoin
def calculate_pearson_correlation_btc(stock_df, stock_name):
    # Merge stock and Bitcoin dataframes on the index (bi-weekly timestamps)
    merged_df = pd.merge(
        stock_df[['close_pct_change']], 
        bi_weekly_crypto_df_btc[['close_pct_change']], 
        left_index=True, 
        right_index=True, 
        suffixes=(f'_{stock_name}', '_btc')
    )
    
    # We have the first line always having a Nan because it is not possible to calculate a % with a close date from before!
    merged_df = merged_df.dropna()

    # Calculate Pearson correlation between the close percentage changes
    if len(merged_df) > 1:
        pearson_corr = merged_df.corr(method='pearson').iloc[0, 1]
        print(f"Pearson correlation between {stock_name} and BTC/USD close percentage changes: {pearson_corr}")
    else:
        print(f"Not enough data points to calculate correlation between {stock_name} and BTC/USD.")

# Calculate Pearson correlations for each stock with Bitcoin
calculate_pearson_correlation_btc(bi_weekly_df_amzn, 'AMZN')
calculate_pearson_correlation_btc(bi_weekly_df_aapl, 'AAPL')
calculate_pearson_correlation_btc(bi_weekly_df_msft, 'MSFT')
calculate_pearson_correlation_btc(bi_weekly_df_nvda, 'NVDA')
calculate_pearson_correlation_btc(bi_weekly_df_meta, 'META')

#Now the same thing with Ethereum! 

# Ethereum data (again, can be ignored but is here to then clean up for the next steps!)
bi_weekly_crypto_df_eth.index = bi_weekly_crypto_df_eth.index.tz_convert('UTC')  # Ensure Ethereum data is in UTC
bi_weekly_crypto_df_eth['close_pct_change'] = bi_weekly_crypto_df_eth['close'].pct_change() * 100

# Function to calculate Pearson correlation for a stock vs Ethereum
def calculate_pearson_correlation_eth(stock_df, stock_name):
    # Merge stock and Ethereum dataframes on the index (bi-weekly timestamps)
    merged_df = pd.merge(
        stock_df[['close_pct_change']], 
        bi_weekly_crypto_df_eth[['close_pct_change']], 
        left_index=True, 
        right_index=True, 
        suffixes=(f'_{stock_name}', '_eth')
    )
    
    # Like with Bitcoin, we need to drop the NaN's to ensure we do not review the changes of the first 
    merged_df = merged_df.dropna()

    # Calculate Pearson correlation between the close percentage changes
    if len(merged_df) > 1:
        pearson_corr = merged_df.corr(method='pearson').iloc[0, 1]
        print(f"Pearson correlation between {stock_name} and ETH/USD close percentage changes: {pearson_corr}")
    else:
        print(f"Not enough data points to calculate correlation between {stock_name} and ETH/USD.")

# Calculate Pearson correlations for each stock with Ethereum
calculate_pearson_correlation_eth(bi_weekly_df_amzn, 'AMZN')
calculate_pearson_correlation_eth(bi_weekly_df_aapl, 'AAPL')
calculate_pearson_correlation_eth(bi_weekly_df_msft, 'MSFT')
calculate_pearson_correlation_eth(bi_weekly_df_nvda, 'NVDA')
calculate_pearson_correlation_eth(bi_weekly_df_meta, 'META')
```

```python
# Now doing the same thing as Pearson now with Spearman

def calculate_spearman_correlation_btc(stock_df, stock_name):

    merged_df = pd.merge(
        stock_df[['close_pct_change']], 
        bi_weekly_crypto_df_btc[['close_pct_change']], 
        left_index=True, 
        right_index=True, 
        suffixes=(f'_{stock_name}', '_btc')
    )

    merged_df = merged_df.dropna()

    if len(merged_df) > 1:
        spearman_corr = merged_df.corr(method='spearman').iloc[0, 1]
        print(f"Spearman correlation between {stock_name} and BTC/USD close percentage changes: {spearman_corr}")
    else:
        print(f"Not enough data points to calculate correlation between {stock_name} and BTC/USD.")

# Calculate Spearman with Bitcoin
calculate_spearman_correlation_btc(bi_weekly_df_amzn, 'AMZN')
calculate_spearman_correlation_btc(bi_weekly_df_aapl, 'AAPL')
calculate_spearman_correlation_btc(bi_weekly_df_msft, 'MSFT')
calculate_spearman_correlation_btc(bi_weekly_df_nvda, 'NVDA')
calculate_spearman_correlation_btc(bi_weekly_df_meta, 'META')

#Same thing now for Ethereum
def calculate_spearman_correlation_eth(stock_df, stock_name):
   
    merged_df = pd.merge(
        stock_df[['close_pct_change']], 
        bi_weekly_crypto_df_eth[['close_pct_change']], 
        left_index=True, 
        right_index=True, 
        suffixes=(f'_{stock_name}', '_eth')
    )
    
    merged_df = merged_df.dropna()

    if len(merged_df) > 1:
        spearman_corr = merged_df.corr(method='spearman').iloc[0, 1]
        print(f"Spearman correlation between {stock_name} and ETH/USD close percentage changes: {spearman_corr}")
    else:
        print(f"Not enough data points to calculate correlation between {stock_name} and ETH/USD.")

calculate_spearman_correlation_eth(bi_weekly_df_amzn, 'AMZN')
calculate_spearman_correlation_eth(bi_weekly_df_aapl, 'AAPL')
calculate_spearman_correlation_eth(bi_weekly_df_msft, 'MSFT')
calculate_spearman_correlation_eth(bi_weekly_df_nvda, 'NVDA')
calculate_spearman_correlation_eth(bi_weekly_df_meta, 'META')
```

```python
#Key Takeaways:
#Weak Monotonic Relationships: In general, the correlations are quite weak, indicating that the tech stocks you're analyzing don't have strong monotonic relationships with Bitcoin or Ethereum.
#Higher Correlations for BTC: The correlations with Bitcoin tend to be slightly higher than with Ethereum, though the difference is not dramatic.
#Spearman vs. Pearson: Spearman is capturing a slightly different aspect of the relationships, particularly with weaker correlations, as it is less sensitive to linear relationships and more robust to non-linear trends and outliers.
```
